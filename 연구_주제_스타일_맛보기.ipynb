🅰️ [데이터 분석] 생활습관 및 우울증/만성질환 분석

예제[데이터 분석] 수면 시간과 우울증 점수의 관계

설명: 숫자로 딱 떨어지는 결과와 통계적 분석을 원한다면 추천합니다.

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. 데이터셋 준비
data_A = {
    '수면시간(시간)': [4, 5, 5.5, 6, 7, 7.5, 8, 9],
    '우울증 점수(10점 만점)': [8, 7, 6, 5, 3, 2, 2, 1]
}
df_A = pd.DataFrame(data_A)

# 2. 데이터 시각화
plt.figure(figsize=(6, 4))
plt.scatter(df_A['수면시간(시간)'], df_A['우울증 점수(10점 만점)'], color='blue', alpha=0.7)
plt.title('A유형: 수면시간에 따른 우울증 점수 분석')
plt.xlabel('수면시간 (시간)')
plt.ylabel('우울증 점수')
plt.grid(True)
plt.show()

print("💡 인사이트: 수면 시간이 길어질수록 우울증 점수가 낮아지는 경향이 확인됩니다.")
```

🅱️ [사례/트렌드 분석] 사회적 보건 이슈 상관관계 분석 (생성형 AI 활용)

예제[사례/트렌드 분석] 생성형 AI를 활용한 사회 지표 트렌드 수집

설명: 챗GPT 등의 AI를 활용해 연도별 출산율과 1인가구 비율 데이터를 수집하고 시각화해 봅니다.

```python
import pandas as pd
import matplotlib.pyplot as plt

print("🤖 [AI 프롬프트 입력]: '최근 6년간 한국의 합계출산율과 1인가구 비율 추이 데이터를 파이썬 딕셔너리 형태로 만들어줘.'\n")
print(">> 🔄 AI가 데이터를 수집 및 생성 중입니다...\n")

# AI가 생성해 준 데이터라고 가정
data_B = {
    '연도': [2018, 2019, 2020, 2021, 2022, 2023],
    '합계출산율': [0.98, 0.92, 0.84, 0.81, 0.78, 0.72],
    '1인 가구 비율(%)': [29.3, 30.2, 31.7, 33.4, 34.5, 35.5]
}
df_B = pd.DataFrame(data_B)

# 데이터 시각화
fig, ax1 = plt.subplots(figsize=(7, 4))
ax1.plot(df_B['연도'], df_B['합계출산율'], color='red', marker='o', label='합계출산율')
ax1.set_xlabel('연도')
ax1.set_ylabel('합계출산율', color='red')

ax2 = ax1.twinx()
ax2.plot(df_B['연도'], df_B['1인 가구 비율(%)'], color='green', marker='s', label='1인가구 비율')
ax2.set_ylabel('1인 가구 비율(%)', color='green')

plt.title('B유형: 연도별 출산율 및 1인가구 비율 트렌드 (AI 수집 데이터)')
plt.show()

print("💡 인사이트: 수집된 데이터 흐름을 통해 사회적 현상 간의 상관관계를 유추할 수 있습니다.")
```

Ⓒ [리뷰 논문] 보건의료 분야 학술적 트렌드 고찰 (생성형 AI 활용)

예제[리뷰 논문] 생성형 AI(딥리서치)를 활용한 문헌 고찰 에이전트 체험

설명: AI가 수백 편의 논문을 검색하고 핵심 트렌드와 연구 공백을 찾아내는 딥리서치 과정을 시뮬레이션해 봅니다.

```python
import time

print("🔍 [AI 딥리서치 에이전트 가동 시작]")
# 연구 주제 설정 (보건학적 관점으로 수정)
topic = "스마트 워치 등 웨어러블 기기를 활용한 만성질환 예방의 보건학적 효과와 과제"
print(f"👉 설정된 타겟 연구 주제: '{topic}'\n")

# 1단계: 검색 시뮬레이션
print("🌐 1단계: 글로벌 보건학/예방의학 DB (PubMed, 한국보건사회연구원 등) 문헌 검색 중...")
time.sleep(2) # 2초 대기
print("   ✔️ 최근 5년 관련 보건학 논문 및 정책 보고서 853편 발견 완료\n")

# 2단계: 필터링 및 분석 시뮬레이션
print("📖 2단계: 초록(Abstract) 기반 핵심 문헌 50편 필터링 및 심층 읽기 진행 중...")
time.sleep(3)
print("   ✔️ 주요 보건 지표(유병률 변화 등), 타겟 인구통계, 정책적 한계점 추출 완료\n")

# 3단계: 종합 요약
print("🧠 3단계: AI 기반 연구 동향 메타 분석 및 보건학적 인사이트 도출 중...")
time.sleep(2)

# 딥리서치 결과 리포트 출력
print("\n==================================================")
print("📝 [AI 딥리서치 종합 요약 리포트]")
print("==================================================")
print("1. 📈 주요 트렌드: 보건의료 패러다임이 '치료'에서 '예방'으로 넘어가면서, 일상생활 속 웨어러블 데이터를 활용해 고혈압/당뇨 등 만성질환을 선제적으로 관리하려는 보건학적 연구가 급증함.")
print("2. 🔑 핵심 키워드: '모바일 헬스(mHealth)', '건강 형평성(Health Equity)', '디지털 리터러시', '예방의학'")
print("3. 💡 연구 공백(Research Gap): 기기를 쉽게 다루는 젊은 층과 달리, 정작 관리가 가장 필요한 고령층의 '디지털 소외 현상'으로 인한 건강 격차(Health Divide)를 다룬 연구가 부족함. 이 정책적 한계점을 짚어주는 리뷰 논문이 가치 있음.")
print("==================================================\n")

print("🎯 실습 팁: 학생들은 AI가 찾아준 '건강 격차'라는 연구 공백(Gap)을 바탕으로, 단순한 기술 소개를 넘어 보건학적 통찰이 담긴 리뷰 논문의 방향성을 쉽게 설정할 수 있습니다!")
```
