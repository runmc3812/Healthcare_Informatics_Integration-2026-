🅰️ [데이터 분석] 생활습관 및 우울증/만성질환 분석

예제[데이터 분석] 수면 시간과 우울증 점수의 관계

설명: 숫자로 딱 떨어지는 결과와 통계적 분석을 원한다면 추천합니다.

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. 데이터셋 준비
data_A = {
    '수면시간(시간)': [4, 5, 5.5, 6, 7, 7.5, 8, 9],
    '우울증 점수(10점 만점)': [8, 7, 6, 5, 3, 2, 2, 1]
}
df_A = pd.DataFrame(data_A)

# 2. 데이터 시각화
plt.figure(figsize=(6, 4))
plt.scatter(df_A['수면시간(시간)'], df_A['우울증 점수(10점 만점)'], color='blue', alpha=0.7)
plt.title('A유형: 수면시간에 따른 우울증 점수 분석')
plt.xlabel('수면시간 (시간)')
plt.ylabel('우울증 점수')
plt.grid(True)
plt.show()

print("💡 인사이트: 수면 시간이 길어질수록 우울증 점수가 낮아지는 경향이 확인됩니다.")
```

🅱️ [사례/트렌드 분석] 사회적 보건 이슈 상관관계 분석 (생성형 AI 활용)

예제[사례/트렌드 분석] 생성형 AI를 활용한 사회 지표 트렌드 수집

설명: 챗GPT 등의 AI를 활용해 연도별 출산율과 1인가구 비율 데이터를 수집하고 시각화해 봅니다.

```python
import pandas as pd
import matplotlib.pyplot as plt

print("🤖 [AI 프롬프트 입력]: '최근 6년간 한국의 합계출산율과 1인가구 비율 추이 데이터를 파이썬 딕셔너리 형태로 만들어줘.'\n")
print(">> 🔄 AI가 데이터를 수집 및 생성 중입니다...\n")

# AI가 생성해 준 데이터라고 가정
data_B = {
    '연도': [2018, 2019, 2020, 2021, 2022, 2023],
    '합계출산율': [0.98, 0.92, 0.84, 0.81, 0.78, 0.72],
    '1인 가구 비율(%)': [29.3, 30.2, 31.7, 33.4, 34.5, 35.5]
}
df_B = pd.DataFrame(data_B)

# 데이터 시각화
fig, ax1 = plt.subplots(figsize=(7, 4))
ax1.plot(df_B['연도'], df_B['합계출산율'], color='red', marker='o', label='합계출산율')
ax1.set_xlabel('연도')
ax1.set_ylabel('합계출산율', color='red')

ax2 = ax1.twinx()
ax2.plot(df_B['연도'], df_B['1인 가구 비율(%)'], color='green', marker='s', label='1인가구 비율')
ax2.set_ylabel('1인 가구 비율(%)', color='green')

plt.title('B유형: 연도별 출산율 및 1인가구 비율 트렌드 (AI 수집 데이터)')
plt.show()

print("💡 인사이트: 수집된 데이터 흐름을 통해 사회적 현상 간의 상관관계를 유추할 수 있습니다.")
```
